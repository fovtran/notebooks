```python
%%latex
** Mathematical formula for the Fourier Series
\begin{equation*}

V_t = \dfrac{a_0}{2} +    \displaystyle \sum_{i=1}^{\infty}[a_i sin(i \omega_0 t) + b_i cos(i \omega_0 t) ] \\ \\

\tiny If a0/2 is constant that's the DC level \\ \\

V_t = \dfrac{a_0}{2} + \displaystyle \sum_{i=1}^{\infty}[a_i sin(i \omega_0 t + \phi_i) ] \\ \\

V_t = \displaystyle \sum_{i=1}^{\infty}\left[\dfrac{sin((2i - 1) \omega_0 t)}{2i - 1} \right] \\ \\

V_t = sin(\omega_0 t) + \dfrac{1}{3} sin(3 \omega_0 t) + \dfrac{1}{5} sin(5 \omega_0 t) + \dfrac{1}{7} sin(7 \omega_0 t) + ... \\ \\

V_t = 0 \cdot sin(\omega_0 t) + 0 \cdot sin(2 \omega_0 t) + sin(3 \omega_0 t) + sin(4 \omega_0 t) \\ \\

\end{equation*}
any_signal(t) = A*sin(f1*t) + B*sin(f2*t) + C*sin(f3*t) ....
```

```python
%%latex
** General fourier series are 
\begin{equation*}

g(x) = a_0 +  \displaystyle \sum_{n=1}^{\infty}[a_n sin(n x) + b_n cos(n x) ] \\ \\

 am not going to do all the math but if you use the orthogonal signal space 
\left\{{1, cos(nx),sin(nx)}\right\} n\in N  \\ \\

a_n = \dfrac{1}{\pi} \displaystyle \int_{-\pi}^{\pi} g(x)cos(n x)\,\mathrm{d}x \\ \\

b_n = \dfrac{1}{\pi} \displaystyle \int_{-\pi}^{\pi} g(x)sin(n x)\,\mathrm{d}x \\ \\

a_0 = \dfrac{1}{2\pi} \displaystyle \int_{-\pi}^{\pi} g(x)\,\mathrm{d}x \\ \\ \\

Derives in \\
g(x) = \dfrac{a_0}{2} +  \displaystyle \sum_{n=1}^{\infty}[a_n sin(n x) + b_n cos(n x)] \\ \\

For the sake of symmetry the a0 \\
a_0 = \dfrac{1}{\pi} \displaystyle \int_{-\pi}^{\pi} g(x)\,\mathrm{d}x \\ \\

V_t = \sum\limits_{i=-\infty}^\infty c_ie^{ji\omega t} \\ \\
\end{equation*}

By applying euler's

\begin{equation*}
e^{j\phi}=\cos\phi+j \sin\phi \\ \\
a_i = c_i + c_{-i}, b_i = j(c_i + c_-i) \\
or a_i = c_i + c_{-i}, b_i = j(c_i + c_{-i}) \\
\end{equation*}

For the special case i=0 you have now two options that looks feasible

Apply the formula ai=ci+ci to be consistent with the cases i0 and you get a0=2c0
Define the coefficient a0 so it doesn't relate to the other coefficents ai.\\

For any ith term with i > 0 if ai = 1 you get a sinusoid that extends from -1 to +1. ie
Peak to peak magnitude of any sine term = 2ai.\\
RMS magnitude of any single sine term is sqrt(2).ai \\
Similarly for bi and cos terms.
\\
If we apply the same convention for DC, for DC to have Vdc = a0 implies a DC peak to peak value of 2a0 (as is the case with ai sine terms).
To maintain a consistent convention we set
Vdc = 1/2.a0\\
```
